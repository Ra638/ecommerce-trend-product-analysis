```python
df['Purchase_Amount'].fillna(df['Purchase_Amount'].median(),inplace= True)
```

```python
df['Rating'].fillna(df['Rating'].mode()[0],inplace=True)
```

```python
Q1 = df['Purchase_Amount'].quantile(0.25)
Q3 = df['Purchase_Amount'].quantile(0.75)
IQR = Q3-Q1 
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
outliers = df[(df["Purchase_Amount"] < lower_bound) | (df["Purchase_Amount"] > upper_bound)]
print(outliers)
```

```python
df["Purchase_Amount"] = df["Purchase_Amount"].clip(lower_bound, upper_bound)
```

```python
plt.figure(figsize = (12,6))
sns.histplot(df['Purchase_Amount'],bins= 30 ,kde= True ,color="blue",alpha = 0.6)
plt.title = ("Distribution of Purchase Amounts")
plt.xlabel = ("Purchase Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Price_per_Item'] = df['Purchase_Amount']/df['Items_Purchased']
df['Estimated_Original_Price'] = df.groupby('Product_Category')['Price_per_Item'].transform("max")*df['Items_Purchased']
df["Estimated_Discount"] = df["Estimated_Original_Price"] - df["Purchase_Amount"]
df.loc[df["Discount_Applied"].isnull(), "Discount_Applied"] = df["Estimated_Discount"]

```

```python
df["Discount_Applied"].isnull().sum()
```

```python



discount_summary = df.groupby(df["Discount_Applied"] > 0)["Purchase_Amount"].agg(["count", "mean", "sum"])

discount_summary.plot(kind="bar", figsize=(8, 5), title="Comparison of Purchases with and without Discounts")
plt.ylabel("Value")
plt.xticks(ticks=[0, 1], labels=["No Discount", "Discount Applied"], rotation=0)
plt.show()
```

```python

# Ensure 'Discount_Applied' is a boolean
df["Discount_Applied"] = df["Discount_Applied"].astype(bool)

# Group by Product Category and Discount Applied
category_performance = df.groupby(["Product_Category", "Discount_Applied"])["Purchase_Amount"].agg(["sum", "count", "mean"]).reset_index()

# Pivot table to see performance with and without discounts
category_pivot = category_performance.pivot(index="Product_Category", columns="Discount_Applied", values="sum").fillna(0)

# Rename columns dynamically
category_pivot.columns = ["No_Discount_Sales", "Discounted_Sales"] if len(category_pivot.columns) == 2 else [f"{col}_Sales" for col in category_pivot.columns]

# Calculate Discount Impact
category_pivot["Discount_Impact"] = category_pivot.get("Discounted_Sales",0) - category_pivot.get("No_Discount_Sales",0)


# Sort by Discount Impact
category_pivot = category_pivot.sort_values(by="Discount_Impact" ,ascending= False)

# Display top 10 underperforming categories
category_pivot.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# Create barplot
sns.barplot(y=category_pivot.index, x=category_pivot["Discount_Impact"], hue=category_pivot.index, palette="coolwarm", legend=False)

# Add labels and title
plt.ylabel("Product Category")

# Show plot
plt.show()
```

🔹 Code to Categorize Customers

```python
customer_spending = df.groupby("Customer_ID")['Purchase_Amount'].sum()
low_threshold = customer_spending.quantile(0.25)
high_threshold = customer_spending.quantile(0.75)
df['Customer_segement'] = df['Customer_ID'].map(
    lambda x : "Low Spender" if customer_spending[x] <= low_threshold else
                "High Spender" if customer_spending[x] >= high_threshold else
                "Medium Spender"
)
df['Customer_segement'].value_counts()
```

🧠 Behavioral KPIs to Analyze

```python
##Average Purchase Value 
total_revenue = df["Purchase_Amount"].sum()
total_orders = len(df)
average_purchase_value = total_revenue/total_orders
print(f"Average Purchase Value (APV):₹{average_purchase_value:.2f}")
```

```python
#Purchase Frequency
total_customers = df["Customer_ID"].nunique()
purchase_frequency = total_orders / total_customers

print(f"Purchase Frequency: {purchase_frequency:.2f} orders per customer")
```

```python
#Repeat Rate 
repeat_customers = df[df.duplicated(subset="Customer_ID",keep=False)]["Customer_ID"].nunique()
repeat_rate = repeat_customers/total_customers
print(f"Repeat Rate: {repeat_rate:.2%}")
```

```python
#customer Lifetime Value 
clv = average_purchase_value*purchase_frequency
print(f"Customer Lifetime Value (CLV): ₹{clv:.2f}")
```

```python
#Return Rate 
total_returns = df['Is_Returned'].sum()
return_rate = total_returns/total_orders
print(f"Return Rate: {return_rate:.2%}")
```

```python

bins = list(range(0, 110, 10))
labels = [f'{i}-{i+10}%' for i in bins[:-1]]
df['Discount_Range'] = pd.cut(df['Estimated_Discount'], bins=bins, labels=labels, include_lowest=True)

heatmap_data = df.pivot_table(values='Is_Returned',
                               index='Rating',
                               columns='Discount_Range',
                               aggfunc='mean')
 
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Return Rate by Rating and Discount Range')
plt.xlabel('Estimated Discount (%)')
plt.ylabel('Customer Rating')
plt.tight_layout()
plt.show()
```

```python

last_purchase = df.groupby('Customer_ID')['Purchase_Date'].max()
returns = df[df['Is_Returned'] == 1][['Customer_ID', 'Purchase_Date']]
returns = returns.merge(last_purchase, on='Customer_ID', suffixes=('_return', '_last'))
churn_customers = returns[returns['Purchase_Date_return'] == returns['Purchase_Date_last']]
churn_rate = len(churn_customers['Customer_ID'].unique()) / df['Customer_ID'].nunique() * 100
print(f"Estimated Churn Rate: {churn_rate:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.pie([churn_rate, 100 - churn_rate], 
        labels=['Churned after Return', 'Others'],
        autopct='%1.1f%%', startangle=140,
        colors=['#ff9999','#66b3ff'])
plt.title('Customer Churn Estimation')
plt.show()
```
